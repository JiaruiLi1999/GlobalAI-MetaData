```python
import pandas as pd
from ib_insync import *
import nest_asyncio
from tqdm.notebook import tqdm
nest_asyncio.apply()
```

```python
ib = IB()
ib.connect('127.0.0.1', 7496, clientId=15)
```

```python
data=pd.read_excel('LIst of 19.7 Stocks with FIGI if available - By Evelyn.xlsx')
data
```

```python
df2=pd.DataFrame(columns = ["longName","marketName","secIdList"])
df3=pd.DataFrame(columns = ["longName","marketName","secIdList"])
```

```python
wrong_name=[]
i_list=[]
```

```python
for i in tqdm(range(len(data))):
    contract = Contract()
    contract.symbol = data.symbol_x[i]
    contract.secType = "STK"
    try:
        bars = ib.reqContractDetails(contract)
        df = pd.DataFrame(bars)
        cleandata = df[["longName", "marketName", "secIdList"]]
        cleandata = cleandata.explode('secIdList')
        cleandata = cleandata.drop_duplicates(subset=["secIdList"])
        
        
        cleandata = cleandata.loc[cleandata.longName==data.longName[i]]
        if len(cleandata.loc[cleandata.longName==data.longName[i]].marketName)==0:
            wrong_name.append(data.symbol_x[i])
        else:
            df2 = pd.concat([df2, cleandata])
        
    except:
        df2 =pd.concat([df2,df3])
        i_list.append(data.symbol_x[i])
    #print(i)
```

```python
df2
```

```python
wrong_data=pd.DataFrame(wrong_name, columns=['Wrong_Name'])
unfound=pd.DataFrame(i_list, columns=['IB_Unfound'])
```

```python
wrong_data.to_csv('wrong_name.csv')
unfound.to_csv('IB_Unfound.csv')
df2.to_csv('isin_16909.csv')
```
