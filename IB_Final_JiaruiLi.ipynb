First Round: Choose Stocks that match the same long name. Get the list of stocks that receive warning, and the lists of stocks that doesn't have correct long name.

```python
for i in tqdm(range(len(data))):
    contract = Contract()
    contract.symbol = data.symbol_x[i]
    contract.secType = "STK"
    try:
        bars = ib.reqContractDetails(contract)
        df = pd.DataFrame(bars)
        cleandata = df[["longName", "marketName", "secIdList"]]
        cleandata = cleandata.explode('secIdList')
        cleandata = cleandata.drop_duplicates(subset=["secIdList"])
        
        
        cleandata = cleandata.loc[cleandata.longName==data.longName[i]]
        if len(cleandata.loc[cleandata.longName==data.longName[i]].marketName)==0:
            wrong_stock.append(data.symbol_x[i])
            wrong_name.append(data.longName[i])
            wrong_country.append(data.Country[i])
            
        else:
            cleandata['Origin_Stock'] = [data.symbol_x[i]]*len(cleandata)
            cleandata['Origin_Name'] = [data.longName[i]]*len(cleandata)
            cleandata['Origin_Country'] = [data.Country[i]]*len(cleandata)
            df2 = pd.concat([df2, cleandata])
        
    except:
        df2 =pd.concat([df2,df3])
        unfound_stock.append(data.symbol_x[i])
        unfound_name.append(data.longName[i])
        unfound_country.append(data.Country[i])
```

Second round: choose stocks that match the ticker name, then get the list of stocks that neither long name is correct or ticker is correct (still_wrong), and list of stocks that has multiple return with correct ticker but incorrect name.

```python
for i in tqdm(range(len(wrong))):
    contract = Contract()
    contract.symbol = wrong.wrong_stock[i]
    contract.secType = "STK"
    try:
        bars = ib.reqContractDetails(contract)
        df = pd.DataFrame(bars)
        cleandata = df[["longName", "marketName", "secIdList"]]
        cleandata = cleandata.explode('secIdList')
        cleandata = cleandata.drop_duplicates(subset=["secIdList"])
        
        
        cleandata = cleandata.loc[cleandata.marketName==wrong.wrong_stock[i]]
        if len(cleandata.loc[cleandata.marketName==wrong.wrong_stock[i]].marketName)==0:
            still_wrong_stock.append(wrong.wrong_stock[i])
            still_wrong_name.append(wrong.wrong_name[i])
            still_wrong_country.append(wrong.wrong_country[i])
        elif len(cleandata.loc[cleandata.marketName==wrong.wrong_stock[i]].marketName)!=1:
            multiple_stock.append(wrong.wrong_stock[i])
            multiple_name.append(wrong.wrong_name[i])
            multiple_country.append(wrong.wrong_country[i])
        else:
            cleandata['Origin_Stock'] = [wrong.wrong_stock[i]]*len(cleandata)
            cleandata['Origin_Name'] = [wrong.wrong_name[i]]*len(cleandata)
            cleandata['Origin_Country'] = [wrong.wrong_country[i]]*len(cleandata)
            df4 = pd.concat([df4, cleandata])
            
        
    except:
        df4 =pd.concat([df4,df3])
```
