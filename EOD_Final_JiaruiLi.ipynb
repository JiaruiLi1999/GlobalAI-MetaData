```python
for j in range(17):
    for i in tqdm(range(1000*j, 1000*j+1000)):
        try:
            newdata = f'https://eodhistoricaldata.com/api/search/{isin_list[i]}?api_token=5f3afd582bd7b4.95720069'
            a = requests.get(newdata).json()
            df1 = pd.DataFrame(a)
            df1['Origin_Stock'] = [data.Origin_Stock.iloc[i]]*len(df1)
            df1['Origin_Name'] = [data.Origin_Name.iloc[i]]*len(df1)
            df1['Origin_Country'] = [data.Origin_Country.iloc[i]]*len(df1)
            first_2_isin(df1)
            country_by_isin(df1)
            df2=df1.loc[df1.Code==df1['Origin_Stock'][0]]

            
            if len(df2)==0:
                wrong_stock_match_ISIN.append(data.secIdList.iloc[i])
                wrong_stock_match_name.append(data.Origin_Name.iloc[i])
                wrong_stock_match_stock.append(data.Origin_Stock.iloc[i])
                
            if len(df2)==1:
                df = pd.concat([df,df2])

            else:
                df3=df2.loc[df2.Country==df2['Origin_Country'][0]]

                if len(df3)==0:
                    df4=df2.loc[df2.Country==df2['country_by_isin'][0]]
                    
                    wrong_country_match_ISIN.append(data.secIdList.iloc[i])
                    wrong_country_match_name.append(data.Origin_Name.iloc[i])
                    wrong_country_match_stock.append(data.Origin_Stock.iloc[i])                    
                    
                    
                    if len(df4)==0:
                        df = pd.concat([df,df2])
                    else:
                        df = pd.concat([df,df4])
    
                else: 
                    df = pd.concat([df,df3])
        except:
            df=pd.concat([df,empty]) 
```
