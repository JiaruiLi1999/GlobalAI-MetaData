```python
l1=ib['isin'].iloc[:5086].tolist()
l2=ib['isin'].iloc[5087:].tolist()
figi['ISIN']=l1+l2
```

```python
figi_split = np.split(figi,[5086])
figi_new=pd.concat([figi_split[0],pd.DataFrame([[np.NaN, 'EOA: E.ON SE: ADJ', 'EOAB']+[np.NaN]*10],columns = figi.columns),figi_split[1]],ignore_index=True)
```

```python
figi_new['new_index']=list(range(len(figi_new)))
ib['new_index']=list(range(len(ib)))
figi_new=figi_new.set_index(['new_index'])
ib=ib.set_index(['new_index'])
ib_figi=pd.concat([ib, figi_new], axis=1).drop(['ISIN',], axis=1)
ib_figi
```
